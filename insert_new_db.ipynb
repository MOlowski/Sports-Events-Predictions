```python
def get_data(table):
    
    conn = None
    params = {
        'host': 'localhost',
        'database': 'preds',
        'user': 'postgres',
        'password': 'pass',
        'port': '5432'
    }
    
    try:
        conn = psycopg2.connect(**params)
    
        query = """
        SELECT *
        FROM {}
        """.format(table)
    
        df = pd.read_sql_query(query, conn)
        
        return df
    except Exception as e:
        return e
    finally:
        if conn is not None:
            conn.close()
```

```python
import psycopg2
import pandas as pd

tables = ['countries', 
          'leagues', 
          'teams',
          'team_stats',
          'fixtures',
          'fixture_stats',
          'players',
          'player_stats',
          'fixtures_updated',
          'predictions']
big_df = {}

df_0 = get_data(tables[0])
df_1 = get_data(tables[1])
df_2 = get_data(tables[2])
df_3 = get_data(tables[3])
df_4 = get_data(tables[4])
df_5 = get_data(tables[5])
df_6 = get_data(tables[6])
df_7 = get_data(tables[7])
df_8 = get_data(tables[8])
df_9 = get_data(tables[9])
```

```python
def insert_table(table_name, df):
    conn = None
    cur = None
    params = {
        'host': 'localhost',
        'database': 'preds',
        'user': 'postgres',
        'password': 'pass',
        'port': '5432'
    }
    try:
        # Establish the connection
        conn = psycopg2.connect(**params)
        cur = conn.cursor()
        
        #insert data into tables
        insert_query = """
            INSERT INTO {} ({})
            VALUES ({})
        """.format(table_name, ','.join(df.columns), ','.join(['%s']*len(df.columns)))
        cur.executemany(insert_query, df.values.tolist())
        print(f'table {table_name} updated')
        
        # Commit the changes
        conn.commit()
        
    except Exception as e:
        print(e)
    finally:
        if conn is not None:
            # Close the cursor and connection
            cur.close()
        if cur is not None:
            conn.close()
```

```python
insert_table('countries', df_0)
insert_table('leagues', df_1)
insert_table('teams', df_2)
insert_table('team_stats', df_3)
insert_table('fixtures', df_4)
insert_table('fixture_stats', df_5)
insert_table('players', df_6)
insert_table('player_stats', df_7)
insert_table('fixtures_updated', df_8)
insert_table('predictions', df_9)
```
